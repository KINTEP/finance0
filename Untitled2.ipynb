```python
def weierstrass(a,b,x, loop = 20):
    nums = len(x)
    w0 = np.zeros(nums)
    for i in range(20):
        a1 = a**i
        b1 = b**i
        ans = b1*np.cos(a1*x*np.pi)
        w0 += ans
    return w0

def sine(time, n, a):
    sx1 = np.zeros(len(time))
    for i in range(n):
        ans = np.sin((a ** i) * np.pi*(time**i))
        sx1 += ans
    return sx1
```

```python
def normalize(sequence):
    max1 = np.max(sequence)
    min1 = np.min(sequence)
    data = []
    for val in sequence:
        norm = (val - min1)/(max1 - min1)
        data.append(norm)
    return data
```

```python
def similarity(x1,x2):
    vals = []
    norm1 = normalize(x1)
    norm2 = normalize(x2)
    for i in range(len(x1)):
        vals.append((norm1[i]+1)/(norm2[i]+1))
    return np.sum(vals)/len(x1)
```

```python

def detrend1(sequence):
    "This takes a sequence and outputs the difference between two successive elements"
    diffs = [0]
    scale = 2
    for i in range(1, len(sequence)):
        prev = i - 1
        now = i
        ans = sequence[prev] - sequence[now]
        fac = int(ans)/np.math.factorial(i)
        diffs.append(fac)
        #scale *= 1
    return diffs

def derivative(data, n):
    derivatives = [data]
    for i in range(n):
        d = detrend1(derivatives[-1])
        derivatives.append(d)
    return derivatives

def derivative2(data, n):
    derivatives = [data]
    for i in range(n):
        d = ratios(derivatives[-1])
        derivatives.append(d)
    return derivatives

def ratios(sequence):
    "This takes a sequence and outputs the difference between two successive elements"
    diffs = [1]
    scale = 1
    for i in range(1, len(sequence)):
        prev = i - 1
        now = i
        dd1 = ((sequence[prev])+1)/((sequence[now]+1))
        diffs.append(dd1/np.math.factorial(scale))
        scale *= scale
    return diffs
```
