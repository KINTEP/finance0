```python
def kernel(m1, m2, l=1):
    return np.exp(- 1 / (2 * l**2) * (m1[:, None] - m2)**2)
```

```python
def standardize(list1):
    std = np.std(list1)
    mean = np.mean(list1)
    standad = []
    for v in list1:
        z = (v - mean)/std
        standad.append(z)
    return standad
```

```python
n = len(f1)
# Domain of f* we want to infer
x_s = np.array([i for i in range(n)])#np.linspace(0, 10, n)

# covariance matrix
K_ss = kernel(x_s, x_s, l=0.2)

# Is square root of matrix (standard deviation)
L_ss = np.linalg.cholesky(K_ss + np.eye(n))

# N~(0, I) * L
f_prior = np.dot(L_ss, np.random.normal(size=(n, 2)))

plt.figure(figsize=(16, 6))
plt.plot(x_s, f_prior)[0]
```

```python
# find mean mu*
K = kernel(x1, x1, l=0.2)
K_s = kernel(x1, x_s, l=0.2)
L = np.linalg.cholesky(K + np.eye(len(x1)))
alpha = np.linalg.solve(L.T, np.linalg.solve(L, f0))
mu = K_s.T @ alpha
```

```python
def dft(x):
    x = np.asarray(x, dtype=float)
    N = x.shape[0]
    n = np.arange(N)
    k = n.reshape((N, 1))
    M = np.exp(-2j * np.pi * k * n / N)
    return np.dot(M, x)
```

```python
def SMA(data, n):
    new_lst = []
    N1 = len(data) - n + 1
    for i in range(N1):
        temp_lst = np.array(data[i:n])
        avg = np.mean(temp_lst)
        new_lst.append(avg)
        n += 1
    return new_lst
```

```python
def fourier(data):
    N = len(data)
    time = np.arange(N)
    Z = np.zeros(N)
    ampS = []
    ampC = []
    freq = []
    for i in range(N):
        f = i
        s1 = np.sin(f*2*np.pi*time/N)
        c1 = np.cos(f*2*np.pi*time/N)
        a1 = np.dot(s1, data)
        b1 = np.dot(c1, data)
        Z += (a1*s1 + b1*c1)
        ampS.append(a1)
        ampC.append(b1)
        freq.append(i)
    Z = Z/N
    return Z, ampS, ampC, freq
```

```python
def forcast(time, a1, b1, f1):
    N = time1.shape[0]
    Z = np.zeros(N)
    for i in range(N):
        s1 = np.sin(f1[i]*2*np.pi*time/N)
        c1 = np.cos(f1[i]*2*np.pi*time/N)
        a11 = a1[i]
        b11 = b1[i]
        Z += (a11*s1 + b11*c1)
    Z = Z/N
    return Z
```

```python
def derivative(a1, b1, time, f):
    N = len(time)
    Z = np.zeros(N)
    for i in range(N):
        c1 = a1[i]*2*np.pi*f[i]/N
        s1 = c1*np.cos(2*np.pi*f[i]*time/N)
        
        c2 = -1*b1[i]*2*np.pi*f[i]/N
        s2 = c2*np.sin(2*np.pi*f[i]*time/N)
    Z += (s1 + s2)
    return Z/N
```

```python
def fourier2(data):
    N = len(data)
    time = np.arange(1, N+1)
    Z = np.zeros(N)
    ampS = []
    ampC = []
    freq = []
    for i in range(N):
        f = i
        s1 = np.sin(f*2*np.pi*time/N)
        c1 = np.cos(f*2*np.pi*time/N)
        a1 = np.dot(s1, data)
        b1 = np.dot(c1, data)
        Z += (a1*s1 + b1*c1)
        ampS.append(a1)
        ampC.append(b1)
        freq.append(i)
    Z = Z/N
    return Z, ampS, ampC, freq
```
