```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def weierstrass(a,b,x, loop = 20):
    nums = len(x)
    w0 = np.zeros(nums)
    for i in range(20):
        a1 = a**i
        b1 = b**i
        ans = b1*np.cos(a1*x*np.pi)
        w0 += ans
    return w0

def sine(time, n, a):
    sx1 = np.zeros(len(time))
    for i in range(n):
        ans = np.sin((a ** i) * np.pi*(time**i))
        sx1 += ans
    return sx1
```

```python
df = pd.read_csv("S&P5001.csv")

seq1 = df.CLOSE[:200]
TIME = np.linspace(0,1,len(seq1))
```

```python
A = np.linspace(5,10,10000000)
B = np.linspace(0,100,10000000)

ch1 = np.random.choice(A)
ch2 = np.random.choice(A)

corrs = []
corrs2 = []
Ass = []
Bss = []
CAss = []
CBss = []
for _ in range(100):
    ch1 = np.random.choice(A)
    ch2 = np.random.choice(A)
    ww1 = weierstrass(a=ch1,b=ch2,x=TIME, loop = 20)
    cum1 = np.cumsum(ww1)
    corr = np.corrcoef(ww1, seq1)[0][1]
    corr2 = np.corrcoef(cum1, seq1)[0][1]
    if abs(corr) >= 0.5:
        Ass.append(ch1)
        Bss.append(ch2)
        corrs.append(corr)
    if abs(corr2) >= 0.5:
        CAss.append(ch1)
        CBss.append(ch2)
        corrs2.append(corr2)
```

```python
def normalize(sequence):
    max1 = np.max(sequence)
    min1 = np.min(sequence)
    data = []
    for val in sequence:
        norm = (val - min1)/(max1 - min1)
        data.append(norm)
    return data
```

```python
seq3 = df.CLOSE[:1000]
```

```python
time11 = np.arange(1,8)
loops = np.arange(4,90)
A_vals = np.linspace(100,300,10000)

totim = np.array([i for i in range(1, len(seq3)+1)])

kora = []
A_select = []
T_select = []
N_select = []

for _ in range(300000):
    t1 = -1 * np.random.choice(time11)
    t2 = np.random.choice(time11)
    n1 = np.random.choice(loops)
    a1 = np.random.choice(A_vals)
    ss1 = sine(time=totim, n=n1, a=a1)
    corr0 = np.corrcoef(np.cumsum(ss1), seq3)[0][1]
    if abs(corr0) >= 0.70:
        kora.append(corr0)
        A_select.append(a1)
        T_select.append([t1, t2])
        N_select.append(n1)
```

```python
def similarity(x1,x2):
    vals = []
    norm1 = normalize(x1)
    norm2 = normalize(x2)
    for i in range(len(x1)):
        vals.append((norm1[i]+1)/(norm2[i]+1))
    return np.sum(vals)/len(x1)
```

```python
similarity(x1=[1,2,3,4],x2=[1,2,3,1])
```

```python
def derivative(data, n):
    derivatives = [data]
    for i in range(n):
        d = detrend(derivatives[-1])
        derivatives.append(d)
    return derivatives

def derivative2(data, n):
    derivatives = [data]
    for i in range(n):
        d = detrend2(derivatives[-1])
        derivatives.append(d)
    return derivatives

def ratios(sequence):
    "This takes a sequence and outputs the difference between two successive elements"
    diffs = [1]
    for i in range(1, len(sequence)):
        prev = i - 1
        now = i
        diffs.append(((sequence[prev])+1)/((sequence[now]+1)))
    return diffs
```
